```python
image_size = 200
```

```python
input_image =[]

for i in cate:
    folders = os.path.join(path, i)
    label = cate.index(i)              # here we need to tell the model that whichy image is of cat and which image is dog.
    print("Value of label is " , label)
    for image in os.listdir(folders):
        image_path = os.path.join(folders, image)
        image_array = cv2.imread(image_path)
        image_array = cv2.resize(image_array , (image_size , image_size))   # Resizing the images to 200 * 200
        input_image.append([image_array, label]) # We are attaching ther target variable
```

```python
len(input_image)
```

```python
np.random.shuffle(input_image)
```

```python
# Till this poinnt X & Y values are tpogether now lets seperate them

X = []
Y = []

for X_values , label in input_image:
    X.append(X_values)
    Y.append(label)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
# Designing the Convolutional Neural Network
```

```python
from keras.models import Sequential
from keras.layers import Conv2D , MaxPool2D , Flatten , Dense , Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters = 10 , kernel_size = (5,5) , activation ="relu" , padding="same", input_shape=(200,200,3)))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Flatten())

model.add(Dense(128, activation ="relu"))
model.add(Dense(2, activation ="softmax"))
```

```python
model.compile(optimizer="adam" , loss = "sparse_categorical_crossentropy" , metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=10 , batch_size = 400 )
```

```python
pred_values = model.predict(X)
pred_values
```

```python
pred_classes = np.argmax(pred_values , axis=1)
pred_classes
```

```python
from sklearn.metrics import confusion_matrix
```

```python
catdog_tab = confusion_matrix(Y , pred_classes)
catdog_tab
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y, pred_classes)*100
```
